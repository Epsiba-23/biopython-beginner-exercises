````python
from Bio.Seq import Seq

# Define DNA sequences using Biopython's Seq class
seq1 = Seq("TGTGACTA") # First DNA sequence
seq2 = Seq("TGGCA")   # Second DNA sequence
seq3 = Seq("CATGGTCA") # Third DNA sequence
````

### Pairwise Alignment Example

This code uses Biopython's `Align` module to perform pairwise alignment between two DNA sequences:

- **Aligner Creation**: A `PairwiseAligner` object is instantiated to align sequences.
- **Performing Alignment**: The aligner is used to align the previously defined DNA sequences, `seq1` ("TGTGACTA") and `seq2` ("TGGCA").
- **Output of Alignments**:alignment result is printed, showing how the two sequences match.
- **Alignment Score**: The score of the alignment is calculated and printed, indicating the quality of the alignment based on the default scoring scheme.



````python
from Bio import Align
aligner = Align.PairwiseAligner()
print(aligner)
alignments = aligner.align(seq1, seq2)
for alignment in alignments:
   print(alignment)  #output of the alignement
print('Score:', aligner.score(seq1,seq2)) #alignment score
````

### Percentage Identity Calculation

below code defines a function to calculate the percentage identity between two aligned sequences i.e quantitative measure of how similar two sequences are after alignment.

- **Function Definition**: The `percent_id` function takes three parameters:
  - `score`: The alignment score obtained from the pairwise alignment.
  - `seq1`: The first DNA sequence.
  - `seq2`: The second DNA sequence.
  
- **Calculation**:
  - The mean length of the two sequences is calculated.
  - The percentage identity is computed using the formula given.


  - The alignment score is obtained using the aligner.
  - The percentage identity is calculated by calling the `percent_id` function.
  - Finally, the percentage identity is printed.


````python
#Let's write a function to calculate percentage identity the formula used
def percent_id(score, seq1, seq2):
  mean_length = (len(seq1) + len(seq2)) / 2
  percentage_identity = (score / mean_length) * 100
  return percentage_identity

score= aligner.score(seq1,seq2) #alignment score
pid= percent_id(score, seq1,seq2) #precentage identity
print("Percentage identity:", pid) #printed percentage
````

### Local Alignment with Custom Gap Scoring

below code snippet demonstrates how to perform local alignment between two DNA sequences:

- **Setting Alignment Mode**: The aligner mode is set to **local**, allowing the identification of the best matching region between the two sequences.
  
- **Gap Scoring**:
  - **Open Gap Score**: Set to -1, indicating the penalty for opening a gap in the alignment.
  - **Extend Gap Score**: Set to -0.5, representing the penalty for extending an existing gap.

- **Aligner Information**: The aligner settings are printed to provide details about the current configuration.

- **Performing Alignment**: The aligner is used to align the sequences `seq1` ("TGTGACTA") and `seq2` ("TGGCA").

- **Output of Alignments**: Each alignment result is printed, showing how the two sequences match locally.

- **Alignment Score**: The score of the local alignment is printed, providing insight into the quality of the alignment based on the specified scoring system.

This example illustrates how to adjust gap penalties in local alignments to better fit specific sequence characteristics.


````python
#local alignmemt
aligner.mode = "local"  #setting alignment mode
aligner.open_gap_score = -1  #open gap score
aligner.extend_gap_score = -0.5  #extend gap score
print(aligner)                    # to get aligner information
alignments = aligner.align(seq1, seq2) #aligner is used to align the two seq
for alignment in alignments:
    print(alignment)             #output of alignment is printed
print(aligner.score(seq1,seq2)) #alignment score
````

### Loading the BLOSUM62 Substitution Matrix

The below code snippet demonstrates how to load and display the BLOSUM62 substitution matrix: highlighting how to work with substitution matrice.

- **Importing Substitution Matrices**: The `substitution_matrices` module from Biopython is imported to access various substitution matrices.

- **Loading Substitution Matrices**: The `load()` function is called to load all available substitution matrices. This is useful for checking the available options.

- **Loading BLOSUM62 Matrix**: Specifically, the BLOSUM62 substitution matrix is loaded using the `load("BLOSUM62")` function. This matrix is commonly used for protein sequence alignment and provides scores for substituting one amino acid for another.

- **Displaying the Matrix**: The BLOSUM62 matrix is printed, showing the substitution scores between amino acids. Positive values indicate favorable substitutions, while negative values indicate unfavorable ones.


````python
from Bio.Align import substitution_matrices #importing substitution matrix
substitution_matrices.load()                #loading substitution matrix
matrix = substitution_matrices.load("BLOSUM62") #loading BLOSUM62 matrix
print(matrix)  #displaying the matrix
````

### Setting the Substitution Matrix for Alignment

The below code snippet demonstrates how to set a custom substitution matrix for the aligner and calculate the alignment score using the BLOSUM62 matrix:

- **Assigning the Substitution Matrix**: The previously loaded BLOSUM62 substitution matrix is assigned to the aligner with `aligner.substitution_matrix = matrix`. This allows the aligner to use the specific scoring scheme defined by the BLOSUM62 matrix for amino acid substitutions during the alignment process.

- **Calculating the Alignment Score**: The score of the alignment between the sequences `seq1` ("TGTGACTA") and `seq2` ("TGGCA") is calculated using the aligner's scoring method. The resulting score reflects the quality of the alignment based on the provided substitution matrix.

````python
aligner.substitution_matrix = matrix
print(aligner.score(seq1,seq2))
````

Another function for pairwise alignment

1. **Importing the `pairwise2` Module**:

   from Bio import pairwise2

2. The pairwise2.align.globalxx() function is called to perform a global alignment of seq1 and seq3.
3.The pairwise2.align.localxx() function performs a local alignment, identifying the most similar regions between seq1 and seq3 without aligning the entire sequences.
4.The format_alignment() function is imported to format the alignment results for better readability.
A loop iterates through each global alignment result, and the formatted alignments are printed to the console.

````python
from Bio import pairwise2
global_alignments = pairwise2.align.globalxx(seq1, seq3)
print(global_alignments)
local_alignments = pairwise2.align.localxx(seq1,seq3)

from Bio.pairwise2 import format_alignment
for alignment in global_alignments:
    print(format_alignment(*alignment))

print("\n***************\n")  #A separator line is printed for clarity between global and local alignment results.

for alignment in local_alignments:
    print(format_alignment(*alignment))
````
